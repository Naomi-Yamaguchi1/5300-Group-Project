```python
import pandas as pd
import numpy as np
```

```python
# import data using pandas

data = pd.read_csv('data/clean_data/formodel_clean_kidney_disease.csv')

data.head()
```

```python
data.shape
```

```python
def categorize_risk(row):
    if row['class'] == 2:
        if row['bp'] < 120 and row['sc'] < 1.4:
            return 'Low Risk'
        else:
            return 'Medium Risk'
    elif row['class'] == 1:
        if row['bp'] >= 140 or row['sc'] > 2.0:
            return 'High Risk'
        else:
            return 'Medium Risk'
        
data['risk_level'] = data.apply(categorize_risk, axis=1)

data.head()
```

```python
data['risk_level'].value_counts()
```

```python
# ckd distribution for new risk level

data.groupby('risk_level')['class'].value_counts()
```

```python
# Remove those columns that derived to create new risk level
data.drop(['bp', 'sc','class'], axis=1, inplace=True)
```

```python
# convert risk_level to numerical
data['risk_level'] = data['risk_level'].map({'Low Risk': 0, 'Medium Risk': 1, 'High Risk': 2})
# change column risk_level to target
data.rename(columns={'risk_level': 'class'}, inplace=True)
# remove first column
data = data.iloc[:,1:]
data.head()
```

```python
# save the data
data.to_csv('data/clean_data/for_model_kidney_disease_multi.csv', index=False)
data.shape
```
